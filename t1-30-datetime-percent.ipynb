# T1-30. 12월 25일 결제 금액(price)은 12월 총 결제금액의 몇 %인가? (정수로 출력)

```python
import pandas as pd
df = pd.read_csv("/kaggle/input/bigdatacertificationkr/payment.csv")
df
```

```python
# your code
df['date'] = pd.to_datetime(df['date'], format='%Y%m%d')
df
```

```python
# 12월 총 결제금액
price12 = df[df['date'].dt.month == 12]['price'].sum()
print(price12)
```

```python
# 12월 25일 결제 금액(price)

price25 = df[df['date'].dt.day == 25]['price'].sum()
print(price25)
```

```python
#  몇 %인가? (정수로 출력)

print(int((price25/price12)*100))
```
