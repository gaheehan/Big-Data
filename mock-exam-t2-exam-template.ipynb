```python
# 심플 결측치 처리
X_train = X_train.fillna('X')
X_test = X_test.fillna('X')
```

```python
print("##Train##")
print(X_train.isnull().sum())
print("##Test##")
print(X_test.isnull().sum())
```

```python
X_test_id = X_test.pop('enrollee_id')
print(X_test_id)
```

```python
X_train = X_train.drop(['enrollee_id','city'], axis=1)
X_test = X_test.drop('city', axis=1)
print(X_train.shape, X_test.shape)
```

```python
# 카테고리 컬럼 
cat_cols = X_train.select_dtypes("object").columns
print(cat_cols)
```

```python
# 라벨 인코딩

from sklearn.preprocessing import LabelEncoder

all_df = pd.concat([X_train.assign(ind='train'),X_test.assign(ind='test')])
le = LabelEncoder()
all_df[cat_cols] = all_df[cat_cols].apply(le.fit_transform)

X_train = all_df[all_df['ind'] == 'train']
X_train = X_train.drop('ind', axis=1)

X_test = all_df[all_df['ind'] == 'test']
X_test = X_test.drop('ind',axis=1)
print(X_test)
```

```python
# 스케일링

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

n_cols = ['city_development_index', 'training_hours']

X_train[n_cols] = scaler.fit_transform(X_train[n_cols])
X_test[n_cols] = scaler.transform(X_test[n_cols])

print(X_train)
```

```python
X = X_train
y = y_train['target']
test = X_test
```

```python
# 데이터 분리
from sklearn.model_selection import train_test_split

X_tr,X_val,y_tr,y_val = train_test_split(X, y, test_size= 0.1, random_state=2023)
print(X_tr.shape,X_val.shape,y_tr.shape,y_val.shape)
```

```python
# 모델학습 및 평가
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_auc_score
rfc = RandomForestClassifier()
rfc.fit(X_tr, y_tr)
pred = rfc.predict_proba(X_val)[:,1]
print(roc_auc_score(y_val, pred))
```

```python
# 최종제출

rfc.fit(X, y)
pred = rfc.predict_proba(test)[:,1]
```

```python
# csv생성코드
submission = pd.DataFrame({'enrollee_id':X_test_id, 'target':pred})
print(submission)

submission.to_csv('00600.csv', index=False)
```
