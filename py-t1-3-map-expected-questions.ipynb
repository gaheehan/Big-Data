## 결측치 처리
-  주어진 데이터에서 결측치가 80%이상 되는 컬럼은(변수는) 삭제하고, 80% 미만인 결측치가 있는 컬럼은 'city'별 중앙값으로 값을 대체하고 'f1'컬럼의 평균값을 출력하세요!
- 데이터셋 : basic1.csv 오른쪽 상단 copy&edit 클릭 -> 예상문제 풀이 시작
- 강의 영상 : https://youtu.be/WqlpqBRn7x4

```python
# 라이브러리 및 데이터 불러오기

import pandas as pd
import numpy as np

df = pd.read_csv("../input/bigdatacertificationkr/basic1.csv")
df.head()
```

```python
# EDA - 결측값 확인(비율 확인)
df.isnull().sum()
```

```python
# 데이터 shape 
df.shape
```

```python
# EDA - 결측비율 확인
df.isnull().sum()/df.shape[0]
```

```python
# 80%이상 결측치 컬럼, 삭제
# f3 컬럼 삭제

print("삭제 전:", df.shape)
df = df.drop(['f3'], axis=1)
print("삭제 후:", df.shape)

# 결측치가 있는 컬럼을 제거하는 2가지 방법
# df.drop(['B', 'C'], axis=1)
# df.drop(columns=['B', 'C'])
```

```python
# 도시 확인
df['city'].unique()
```

```python
# 80%미만 결측치 컬럼, city별 중앙값으로 대체
s = df[df['city']=='서울']['f1'].median()
k = df[df['city']=='경기']['f1'].median()
d = df[df['city']=='대구']['f1'].median()
b = df[df['city']=='부산']['f1'].median()

s,k,d,b
```

```python
# 대체 전 데이터 샘플 출력
df[18:21]
```

```python
# f1결측치 city별 중앙값으로 대체
df['f1'] = df['f1'].fillna(df['city'].map({'서울':s, '경기':k, '대구':d, '부산':b}))
```

```python
# 대체 후 데이터 샘플 출력
df[18:21]
```

```python
# f1 평균값 결과 출력
print(df['f1'].mean())
```
