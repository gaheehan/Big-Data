## 주어진 데이터는 고혈압 환자 치료 전후의 혈압이다. 해당 치료가 효과가 있는지 대응(쌍체)표본 t-검정을 진행하시오
- 귀무가설(H0): $\mu$ >= 0
- 대립가설(H1): $\mu$ < 0
- $\mu$ = (치료 후 혈압 - 치료 전 혈압)의 평균
- 유의수준: 0.05

1. $\mu$의 표본평균은?(소수 둘째자리까지 반올림)
2. 검정통계량 값은?(소수 넷째자리까지 반올림)
3. p-값은?(소수 넷째자리까지 반올림)
4. 가설검정의 결과는? (유의수준 5%)

```python
import pandas as pd

df = pd.read_csv("/kaggle/input/bigdatacertificationkr/high_blood_pressure.csv")
df.head()
```

```python
# 1.𝜇의 표본평균은?(소수 둘째자리까지 반올림) # -6.12


df['diff'] = df['bp_post'] - df['bp_pre']
d_mean = df['diff'].mean()
print(round(d_mean,2))   
```

```python
# 검정통계량 값은?(소수 넷째자리까지 반올림)  # 3.0002

from scipy import stats

result = stats.ttest_rel(df['bp_pre'], df['bp_post'], alternative = 'greater')
print(round(result.statistic,4))
```

```python
# p-값은?(소수 넷째자리까지 반올림  # 0.0016

print(round(result.pvalue,4))
```

```python
# 가설검정의 결과는? (유의수준 5%) 0.0016 < 0.5 이므로 대립가설 채택, 귀무가설 기각
```

```python
import pandas as pd
df = pd.read_csv("/kaggle/input/bigdatacertificationkr/high_blood_pressure.csv")
df.head()
```

```python
# 𝜇의 표본평균은?(소수 둘째자리까지 반올림)

df['diff'] = df['bp_post'] - df['bp_pre']
d_mean = round(df['diff'].mean(),2)
print(d_mean)  # -6.12
```

```python
# 검정통계량 값은?(소수 넷째자리까지 반올림)

from scipy import stats 

result = stats.ttest_rel(df['bp_pre'], df['bp_post'], alternative = 'greater')
print(result)

print(round(result.statistic,4)) # 3.0002
```

```python
# p-값은?(소수 넷째자리까지 반올림)

print(round(result.pvalue,4))
```

```python
# 가설검정의 결과는? (유의수준 5%)
# 귀무가설 기각, 대립가설 채택 
```
