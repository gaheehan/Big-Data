## 나이 구간 나누기
###  basic1 데이터 중 'age'컬럼 이상치를 제거하고, 동일한 개수로 나이 순으로 3그룹으로 나눈 뒤 각 그룹의 중앙값을 더하시오
### (이상치는 음수(0포함), 소수점 값)
- data: basic1.csv
- 오른쪽 상단 copy&edit 클릭 -> 예상문제 풀이 시작


```python
# 라이브러리 및 데이터 불러오기
import pandas as pd

df = pd.read_csv('../input/bigdatacertificationkr/basic1.csv')
df.head()
```

```python
# 'age'컬럼 이상치 제거 (이상치는 음수(0포함), 소수점 값)

print("전체 데이터: ", df.shape)
df = df[~(df['age'] <= 0)]
print("음수(0포함)값 제거 후 데이터: ",df.shape)

df = df[(df['age'] == round(df['age'],0))]
print("소수점 값 제거 후 데이터: ",df.shape)
```

```python
# 기준 확인
pd.qcut(df['age'], q=3)
```

```python
# 구간 분할

df['range'] = pd.qcut(df['age'], q=3, labels = ['group1', 'group2','group3'])
df.head()
```

```python
# 수량비교

df['range'].value_counts()
```

```python
# 각 그룹의 중앙값을 더하시오

g1_m = df[df['range'] == 'group1']['age'].median()
g2_m = df[df['range'] == 'group2']['age'].median()
g3_m = df[df['range'] == 'group3']['age'].median()

print(g1_m + g2_m + g3_m)
```
